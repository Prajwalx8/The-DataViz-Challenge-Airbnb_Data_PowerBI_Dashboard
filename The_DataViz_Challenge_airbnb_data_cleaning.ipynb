# About the Dataset

* Some columns like neighbourhood_group and license may not be useful for analysis.



# **Data cleaning and transformation:**

```python
# Drop unnecessary columns
chicago_and_Orlean.drop(['neighbourhood_group', 'license'], axis=1, inplace=True, errors='ignore')
```

```python
# Remove rows where price is missing
chicago_and_Orlean.dropna(subset=['price'], inplace=True)
```

```python
# Fill missing host_name with most frequent value
chicago_and_Orlean['host_name'] = chicago_and_Orlean['host_name'].fillna(chicago_and_Orlean['host_name'].mode()[0])
```

```python
# Convert last_review to datetime
chicago_and_Orlean['last_review'] = pd.to_datetime(chicago_and_Orlean['last_review'], errors='coerce')
```

```python
# Fill missing last_review with median date
chicago_and_Orlean['last_review'] = chicago_and_Orlean['last_review'].fillna(chicago_and_Orlean['last_review'].median())
```

```python
# Fill missing reviews_per_month with mean
chicago_and_Orlean['reviews_per_month'] = chicago_and_Orlean['reviews_per_month'].fillna(chicago_and_Orlean['reviews_per_month'].mean())
```

```python
print("\nMissing values after cleaning:")
print(chicago_and_Orlean.isnull().sum())
```

# Data Cleaning and Transformation briefing

* **Column Optimization:**
Irrelevant and redundant columns such as neighbourhood_group and license were removed to streamline the dataset and focus on meaningful analytical features.

* **Handling Missing Values** :
Rows with missing price values were removed, as price is a key variable for analysis and visualization.

* **Categorical Data Imputation:**
Missing values in the host_name column were filled using the most frequent (mode) value to maintain consistency in host-related analysis.

* **Date Processing and Imputation:**
The last_review column was converted into a datetime format. Missing values were then imputed using the median date to preserve temporal continuity without introducing bias.

* **Numerical Data Imputation:**
Missing values in reviews_per_month were replaced with the mean value to ensure completeness while maintaining overall data distribution.
